```python
from transformers import pipeline
```

```python
question_answerer = pipeline("question-answering", model='distilbert-base-cased-distilled-squad')
```

```python
context = r"""
Manik Kamra is 26 years old male who loves to travel and have visited 12 countires. He lives in india and he currently works in ABC compnay
and has also experience in Podcasting.
"""
```

```python
result = question_answerer(question="Where does Manik work?",context=context)
print(f"Answer: '{result['answer']}', score: {round(result['score'], 4)}")
```

```python
result = question_answerer(question="How many countries manik has visited till now ?",context=context)
print(f"Answer: '{result['answer']}', score: {round(result['score'], 4)}")
```
